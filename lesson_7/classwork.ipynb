1. Напишите функцию `geom(...)`, которая принимает неограниченное количество неименованных аргументов и возвращает корень из их произведения.

```python
def geom(*numbers):
    result = 1
    
    for number in numbers:
        result *= number
        
    # 1 * numbers[0] * numbers[1]....
    
    return result ** 0.5

geom(2, 3, 6)
```

```python
# def replace_words(text, A, B):
# def replace_words(text, Мама, раму):
# replace_words(text, Мама="sdfsdf", раму="sdfsdf")

def replace_words(text, **replace_pairs):
    print("replace_pairs", replace_pairs)
    print("replace_pairs.items():", replace_pairs.items())
    for old_word, new_word in replace_pairs.items():
        text = text.replace(old_word, new_word)
    return text

replace_words("Мама мыла раму", Мама="Рама", раму="маму")
```

3. Напишите функцию `censor(...)`, которая принимает текст и букву, либо только текст (буква по умолчанию - а) и заменяет в нем все слова с заданной буквой на звездоки. Циклы использовать нельзя.  

```python
%%timeit
def censor(text, letter='а'):
    words = map(lambda word: word if letter not in word.lower()
                else '*' * len(word), text.split())
    return ' '.join(words)

r = censor("Мама мыла булку" * 1000, "у")
```

```python
def proccess_word(word, letter):
    if letter in word.lower():
        return '*' * len(word)
    else:
        return word

def censor(text, letter='а'):
    words = map(lambda word: proccess_word(word, letter), text.split())
    return ' '.join(words)

print(censor("Мама мыла булку"))
print(censor("Мама мыла булку", "у"))
```

```python
def proccess_word(word, letter):
    if letter in word.lower():
        return '*' * len(word)
    else:
        return word

def censor(text, letter='а'):
    words = text.split()
    
    for i in range(len(words)):
        words[i] = proccess_word(words[i], letter)
    
    return ' '.join(words)

print(censor("Мама мыла булку"))
print(censor("Мама мыла булку", "у"))
```

4. Будем считать слово тем интереснее, чем больше в нем процент согласных букв. Напишите функцию `interesting(...)`, которая принимает текст и число N (либо только текст, тогда N  - 5) и возвращает список из N самых интересных слов в тексте.

```python
def interesting(text, N=5):
    vowels = ['а', 'у', 'е', 'ы', 'о', 'э', 'я', 'ю', 'ё', 'и']
    is_vowel = lambda letter:letter in vowels
    count_vowels = lambda word: len(list(filter(is_vowel, word))
    percent_of_vowels = lambda word: count_vowels(word) / len(word)
    
    return sorted(text.split(), key=percent_of_vowels)[:N]

interesting("Мама мгновенно и качественно мыла дубовую раму, а папа читал газету")
```

```python
%%timeit
def fact(n):
    result = 1
    
    for i in range(n + 1):
        result *= i
        
    return result

[fact(i) for i in range(100)]
```

```python
%%timeit
def fact(n):
    # print("Вычисляю факториал {}".format(n))
    
    if n == 0:
        #print("Это 1")
        return 1
    else:
        #print("Это {} умножить на факториал {}".format(n, n - 1))
        return n * fact(n - 1)

[fact(i) for i in range(100)]
```

```python
%%timeit
def fib(n):
    a, b = 0, 1
    
    for i in range(n):
        a, b = b, a + b
    
    return a

[fib(i) for i in range(20)]
```

```python
%%timeit

results = {}

def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    if n in results:
        return results[n]
        
    res = fib(n - 1) + fib(n - 2)
    results[n] = res
    return res

[fib(i) for i in range(20)]
```

```python
def SumLists(lists):
    if isinstance(lists, int):
        return lists
    else:
        return sum(map(SumLists, lists))
```

```python
def JoinLists(lists):
    if isinstance(lists, int):
        return [lists]
    else:
        S = []
        for elem in lists:
            S += JoinLists(elem)
        return S

JoinLists([1, [2, 3], [[4, 5], 6]])
```
