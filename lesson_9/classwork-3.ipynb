```python
import requests 

answer = requests.get("https://nplus1.ru/news/")
```

```python
from bs4 import BeautifulSoup

page = BeautifulSoup(answer.text, 'lxml')
```

```python
articles = page.find_all('article', {'class': 'item'})
```

```python
len(articles)
```

```python
def get_article_content(url):
    answer = requests.get("https://nplus1.ru" + url)
    page = BeautifulSoup(answer.text, 'lxml')
    text = page.find('div', {'class': 'body'})
    
    return text.get_text()
```

```python
import os

os.mkdir('articles')
```

```python
def save_to_file(title, date, text):
    with open("articles\\{} {}.txt".format(date, title), 'w', encoding='utf-8') as file:
        file.write(text)
```

```python
from tqdm.notebook.tqdm import tqdm
```

```python
for article in tqdm(articles):
    title = article.find('h3').get_text()
    link = article.find('a').get('href')
    date_parts = link.split('/')
    date = '-'.join(date_parts[2:5])
    # date = date_parts[2] + '-' + date_parts[3] + '-' + date_parts[4]
    
    text = get_article_content(link)
    save_to_file(title, date, text)
    
    
```
